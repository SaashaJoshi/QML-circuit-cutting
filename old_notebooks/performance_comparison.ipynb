```python
import matplotlib.pyplot as plt
import numpy as np
```

# Compare all performance

```python
# Time to perform one forward pass [CPU (qasm_simulator), CPU (AerSimulator), GPU, AWS Local, AWS SV1, AWS TN1]
forward_sub_A = [1.69, 1.93, 4.78, 6.88, 6.87, 7]
backward_sub_A = [35.8, 40, 97.26, 161, 157.61, 156]

forward_sub_B = [2.96, 3.13, 6.34, 25.08, 25.07, 26]
backward_sub_B = [73.23, 89, 133.28, 593.58, 589.26, 580]
```

```python
# Plotting Forward Sub on one graph
fig, ax_forward = plt.subplots()
ax_forward.plot(range(6), forward_sub_A, 'o-', label='Forward pass for sub-circuits A', color='mediumorchid')
ax_forward.plot(range(6), forward_sub_B, 'o-', label='Forward pass for sub-circuits B', color='salmon')

ax_forward.set_xticks(range(6))
ax_forward.set_xticklabels(['CPU \n(qasm_simulator)', 'CPU \n(AerSimulator)', 'GPU', 'Local', 'SV1', 'TN1'])
ax_forward.set_xlabel('Devices')
ax_forward.set_ylabel('Time (in seconds)')
ax_forward.set_title('Time taken to complete one Forward pass')
ax_forward.legend()
ax_forward.set_ylim(-2, 35)  # Adjust the y-axis limits based on your data range

# Annotate values on the chart
for i, value in enumerate(forward_sub_A):
    ax_forward.text(i, value - 2, f'{value:.2f}', ha='center', va='bottom', color='black', fontsize=10)

for i, value in enumerate(forward_sub_B):
    ax_forward.text(i, value + 1.5, f'{value:.2f}', ha='center', va='bottom', color='black', fontsize=10)

# Increase margin
plt.subplots_adjust(left=0.01, right=1, top=0.9, bottom=0.1)

plt.show()
```

```python
# Plotting Backward Sub on another graph
fig, ax_backward = plt.subplots()
ax_backward.plot(range(6), backward_sub_A, 'o-', label='Backward pass for sub-circuits A', color='mediumorchid')
ax_backward.plot(range(6), backward_sub_B, 'o-', label='Backward pass for sub-circuits B', color='salmon')

ax_backward.set_xticks(range(6))
ax_backward.set_xticklabels(['CPU \n(qasm_simulator)', 'CPU \n(AerSimulator)', 'GPU', 'Local', 'SV1', 'TN1'])
ax_backward.set_xlabel('Devices')
ax_backward.set_ylabel('Time (in seconds)')
ax_backward.set_title('Time taken to complete one Backward pass')
ax_backward.legend()
ax_backward.set_ylim(-10, 650)  # Adjust the y-axis limits based on your data range

# Highlight background for each axis
# ax_backward.axvspan(-0.25, 0.25, facecolor='lightgreen', alpha=0.3)
# ax_backward.axvspan(0.75, 1.25, facecolor='lightblue', alpha=0.3)

# Annotate values on the chart
for i, value in enumerate(backward_sub_A):
    ax_backward.text(i, value - 40, f'{value:.2f}', ha='center', va='bottom', color='black', fontsize=10)

for i, value in enumerate(backward_sub_B):
    ax_backward.text(i, value + 15, f'{value:.2f}', ha='center', va='bottom', color='black', fontsize=10)

# Increase margin
plt.subplots_adjust(left=0.01, right=1, top=0.9, bottom=0.1)

plt.show()
```

```python
# Data
iterations = [5, 10, 50]
CPU = [8.6 + 22.7, 11.63 + 26.21, 28.73 + 69.12]
GPU = [26.65 + 32.88, 32.22 + 40.5, 76.15 + 96.24]

# Plotting
plt.plot(iterations, CPU, marker='o', label='CPU')
plt.plot(iterations, GPU, marker='o', label='GPU')

# Adding labels and title
plt.xticks(range(0, iterations[-1] + 1, 5))
plt.xlabel('Iterations')
plt.ylabel('Time to Train (minutes)')
plt.title('Time to Train 8-qubit QTN with 1 Circuit Cut \n on CPU and GPU for Different Iterations')

# Adding legend
plt.legend()

# Adding annotations
for i, (cpu_time, gpu_time) in enumerate(zip(CPU, GPU)):
    plt.text(iterations[i], cpu_time + 5, f'{cpu_time:.2f} s', ha='right', va='bottom', color='black', fontsize=10)
    plt.text(iterations[i], gpu_time + 10, f'{gpu_time:.2f} s', ha='right', va='top', color='black', fontsize=10)

# Increase margin
plt.xlim(0, 52)
plt.ylim(20, 195)
plt.subplots_adjust(left=0, right=1, top=0.8, bottom=0.1)

# Display the plot
plt.show()
```
